# string


```python
# create string
name = "Python"
print(name)
```

```python
# access 1st index element
greet = 'hello'
print(greet[1]) # "e"
```

```python
# assign new string to message variable
message = 'Hola Amigos'
message = 'Hello Friends'
print(message)
```

```python
# multiline string 
message = """
Never gonna give you up
Never gonna let you down
"""
print(message)
```

```python
# compare
str1 = "Hello, world!"
str2 = "I love Swift."
str3 = "Hello, world!"
print(str1 == str2)
print(str1 == str3)
```

```python
# Join Two or More Strings
greet = "Hello, "
name = "Jack"
result = greet + name
print(result)
```

```python
# iterating through greet string
greet = 'Hello'
for letter in greet:
    print(letter)
```

```python
# String Membership Test
print('a' in 'program')
print('at' not in 'battle')
```

```python
# convert message to uppercase
message = 'python is fun'
print(message.upper())
```

```python
# check if the message starts with Python
message = 'Python is fun'
print(message.startswith('Python'))
```

```python
# replace 'ba' with 'ro'
text = 'bat ball'
replaced_text = text.replace('ba', 'ro')
print(replaced_text)
```
