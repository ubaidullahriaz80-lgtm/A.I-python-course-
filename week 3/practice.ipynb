```python
a = [8, 77, 99, 33, 22, 44]
total = 0
for num in a:
    total += num 
average = total / len(a)
print("Sum:", total)
print("Average:", average)
```

```python
customer_order = [
    'mango','dates','apples',
    'pine apples','apples','dates',
    'mango','dates','apples'
]
unique_items = set(customer_order)
order_count = {}
for item in customer_order:
    if item in order_count:
        order_count[item] += 1
    else:
        order_count[item] = 1
print("Unique Items:", unique_items)
print("Order Counts:")
for item, count in order_count.items():
    print(f"{item}: {count}")
```

# list comphension

```python
number = [1,2,3,4,5]
square_num = [num*num for num in number ]
print(square_num)
```

```python
even_numbers = [num for num in range(1, 20) if num % 2 == 0 ]
print(even_numbers)
```

```python
word = "i love python"
vowels = "aeiou"
result = [char for char in word if char in vowels]
print(result)
```

```python
multiplication = [[i * j for j in range(1, 6)] for i in range(2, 5)]

print(multiplication)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)
```
