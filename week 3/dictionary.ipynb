# Dictionary 

```python
# create dictionary
student1 = {'name':'ali','age':18,'city':'multan','class':9}
print(student1)
```

```python
# adding value in dictionary 
country_capitals = {"Germany": "Berlin","Canada": "Ottawa",}
country_capitals["Italy"] = "Rome"

print(country_capitals)
```

```python
# delete value
country_capitals = {"Germany": "Berlin","Canada": "Ottawa", }
del country_capitals["Germany"]
print(country_capitals)
```

```python
# clear the dictionary
country_capitals = {"Germany": "Berlin","Canada": "Ottawa",}
country_capitals.clear()
print(country_capitals)
```

```python
# change the value of "Italy" key to "Rome"
country_capitals = {"Germany": "Berlin","Italy": "Naples","England": "London"}
country_capitals["Italy"] = "Rome"
print(country_capitals)
```

```python
# Iterate Through a Dictionary
country_capitals = {"United States": "Washington D.C.","Italy": "Rome"}
# print dictionary keys one by one
for country in country_capitals:
    print(country)
print()
# print dictionary values one by one
for country in country_capitals:
    capital = country_capitals[country]
    print(capital)
```
